```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1 = np.array(([1, 2, 3], [4, 5, 6]))

arr2 = np.random.rand(3 ,3) *10

arr3 = np.zeros((4, 4))
```

```python
np.save('arr1.npy', arr1)
np.save('arr2.npy', arr2)
np.save('arr3.npy', arr3)
```

```python
loaded_arr1  = np.load('arr1.npy')
loaded_arr2  = np.load('arr2.npy')
loaded_arr3  = np.load('arr3.npy')
print(loaded_arr1)
print(loaded_arr2)
print(loaded_arr3)
```

```python
try:
    logo = np.load("numpy-logo.npy")
    #Display
    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy logo")
    plt.grid("False")
# dark mode 
    dark_logo = 1 - logo
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Numpy logo")
    plt.grid("False")

except FileNotFoundError:
    print("npy logo file not found  ")
```
