```python
import numpy as np
import  matplotlib.pyplot as plt
```

```python
# Data Structure  : {restaurant_id, 2021, 2022, 2023, 2024}

sales_data = np.array([
    [1, 15000, 18000, 220000, 25000], # Paradise Biryani
    [2, 12000, 14000, 160000, 19000], # Biryani Bites
    [3, 20000, 23000, 260000, 30000], # Pizza hub
    [4, 18000, 21000, 240000, 27000], # Burger point
    [5, 16000, 18500, 2050000, 230000], # Chai Pointes
])

print(" === Zomato sales analysis ===")
print("\n Sales data shape ", sales_data.shape)
# print("\n Sample data for 1st 3 restourent ", sales_data[0:3])
print("\n Sample data for 1st 3 restourent ", sales_data[:, 1:])
```

```python
#  total sales per year

sales_data1 = np.array([
    [1, 15000, 18000, 220000, 10000], # Paradise Biryani
    [2, 12000, 14000, 160000, 19000], # Biryani Bites
    [3, 20000, 23000, 260000, 30000], # Pizza hub
    [4, 18000, 21000, 240000, 27000], # Burger point
    [5, 16000, 18500, 2050000, 230000], # Chai Pointes
])


print(np.sum(sales_data,axis=0))

yearly_total = np.sum(sales_data1[:, 1:], axis=0)
print(yearly_total)

# Minimum sales per restaurant
min_sales = np.min(sales_data1[:, 1:], axis=1)
print(min_sales)

#maximum sales per year

max_sales = np.max(sales_data1[:, 1:], axis=0)
print(max_sales)

# Average sales per restaurent

avg_sales = np.mean(sales_data1[:, 1:], axis=1)
print("Average Sales ", avg_sales)

# Cumulative

cusum = np.cumsum(sales_data1[:, 1:], axis=1)
print(cusum)

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cusum, axis=0))
plt.title("Average cumulative sales across all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
